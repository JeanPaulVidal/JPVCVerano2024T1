```python
datos = [
  'Ramirez','Garcia','Gonzalez','Rodriguez','Perez','Martinez','Garcia','Lopez','Gutierrez',
'Lopez','Rodriguez','Martinez','Rodriguez','Hernandez','Garcia','Hernandez','Lopez',
'Perez','Gonzalez','Martinez','Garcia','Ramirez','Gonzalez','Hernandez','Garcia','Gutierrez',
'Garcia','Rodriguez','Hernandez','Sanchez','Hernandez','Gonzalez','Rodriguez','Sanchez','Hernandez',
'Gonzalez','Gutierrez','Martinez','Gonzalez','Rodriguez','Rodriguez','Martinez','Ramirez',
'Perez','Sanchez','Hernandez','Hernandez','Gonzalez','Garcia','Hernandez','Rodriguez','Lopez',
'Sanchez','Rodriguez','Ramirez','Gonzalez','Lopez','Martinez','Lopez','Garcia','Lopez','Sanchez',
'Martinez','Garcia','Garcia','Hernandez','Garcia','Hernandez','Flores','Perez','Gonzalez','Gonzalez'
,'Lopez','Martinez','Gonzalez','Martinez','Hernandez','Ramirez','Lopez','Sanchez','Ramirez','Gutierrez'
,'Sanchez','Flores','Garcia','Martinez','Garcia','Flores','Perez','Lopez','Hernandez','Martinez','Garcia'
,'Hernandez','Flores','Perez','Gonzalez','Hernandez','Sanchez','Martinez'
]

valores_ref_eje = list(range(1, len(datos) + 1))

clases, fabs = car.frec_abs(datos)
clases_sorted = jen.get_clases(clases)
fabs_sorted = mar.sort_clases_fa(clases, clases_sorted, fabs)
frs = ed.frecRel(fabs_sorted)
facumuladas = iv.calcular_frecuencia_acumulada(frs)


# Imprimir tabla
def printTable(encabezados, contenido):
    from tabulate import tabulate
    if(len(encabezados) == len(contenido)):
        data = list(zip(*contenido))
        table = tabulate(data, headers=encabezados, tablefmt="orgtbl")
        print(table)

printTable(["Clase", "Fa", "Fr", "Fr Ac"], [clases_sorted, fabs_sorted, frs, facumuladas])

'''-------------------------------               -----------------------------'''
'''------------------------------DIAGRAMA DE BARRAS---------------------------'''
'''-----------------------------                   ---------------------------'''
def create_bar_chart(frecuencias, valores_ref_eje):
    # Colores
    colores = ['#FF5733', '#80C0FF', '#FFC080', '#80FF80', '#C080FF'] * (len(valores_ref_eje) // 5 + 1)

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.barh(valores_ref_eje, frecuencias, height=0.8, edgecolor="k", color=colores[:len(valores_ref_eje)])

    ax.invert_yaxis()  # Invertir el sentido del eje y
    ax.set_yticks(range(len(valores_ref_eje)))
    ax.set_yticklabels(valores_ref_eje, fontsize=10)  # Etiquetar marcas de clase
    ax.set_xlabel("Frecuencia absoluta", fontsize=15)
    ax.set_ylabel("Apellidos", fontsize=15)
    ax.set_title("Diagrama de barras de frecuencia de apellidos", fontsize=20)
    ax.grid()

    return fig, ax

'''-------------------------------               -----------------------------'''
'''----------------------------IMPRIMIR GRAFICA DE BARRAS---------------------'''
'''-----------------------------                   ---------------------------'''

def imprimir_grafica(frecuencias, valores_ref_eje):
    fig, ax = create_bar_chart(frecuencias, valores_ref_eje)
    plt.show()

# Example usage
frecuencias = fabs_sorted
valores_ref_eje = clases_sorted
imprimir_grafica(frecuencias, valores_ref_eje)
print("La grfica presenta la frecuencia de aparicin de los diferentes apellidos.")
```
